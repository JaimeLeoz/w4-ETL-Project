```python
#vamos a crear la nueva columna con datos de los rangos en m2 basado en el numero de habitaciones
```

```python
df = df_rmv_columns
```

```python
import numpy as np
```

```python
conditions = [
    (df['bedrooms'] <= 2),
    (df['bedrooms'] == 3),
    (df['bedrooms'] >= 4) & (df['bedrooms'] <= 5),
    (df['bedrooms'] > 5)]

values = ['50', '85', '120', '180']
df['sq_meters'] = np.select(conditions, values)
```

```python
#ha incluido la nueva columna, con nuestras condiciones ( Impresionante)
df.shape
```

```python
#comprobamos si es correcto y lo es!

df[['bedrooms', 'sq_meters']]
```

```python
# ahora haremos unos nuevos condicionales, ya que tenemos las fechas de cuando se escrapearon los arbnb podemos incluir la media de precios de esos meses que
# provee Idealista

# tengo precio por noche, metros cuadrados, podemos generar nueva columna con el calculo si son un mes x metro cuadrado
# 
#  Luego incluimos precios alguiler de mercado
```

```python
df.columns
```

```python
# vamos a generar la columna nueva , precio arbnb x mes. Multiplicamos el precio x 30 noches
# tenemos el $ en la columna price, hay que eliminarlo y cambiar la columna a integer antes de esta formuala
```

```python
df.dtypes
```

```python
#here im replacing the $ sign for nothin and also the "," in the thousands which was not letting me to change the column from a string to a float
# this way now i'm able do the next calculations without the error " cant multiply a string vs integer"
df['price'] = df['price'].str.replace('$','')
df['price'] = df['price'].str.replace(',','')
print(df.price)
```

```python
# downcasting from object string to float the price column

df["price"] = pd.to_numeric(df["price"], downcast="float")
```

```python
# confirming that has worked
print(df.dtypes)
```

```python
# At first iw as goinggto divide price/ by minimum_nights ( because in some cases, the minmum stay can be 3 days, which will make our data unconsistent) then we multiply by 30 days
# but i double check and always price is per night so no need for the division with the minimun night.
# now lets see if we can do the new column with the calculation and make sure it did work

df['price_monthly'] = (df['price']) * 30
df.head()
```

```python
# calculations are fine and consistent

df[['price', 'minimum_nights', 'price_monthly','sq_meters','listing_url']]
```

```python
df.shape
```

```python
# we need to change the column sqm into a integer befor calculating anything
df["sq_meters"] = pd.to_numeric(df["sq_meters"], downcast="integer")
```

```python
# alright! now I will just do a new column with the price / sq_meter _ this output will be to match it with other datasources..
# ..that provide the avg price x sqm of rents by Municipios AKA Districts ( based on Montly data)

df['price_sqm_amonth'] = df['price_monthly'] / df['sq_meters']



```

```python
# looking gorgeous

df[['price', 'minimum_nights', 'price_monthly','sq_meters','listing_url','price_sqm_amonth','last_scraped' ]]
```

```python
# to work in the future with sql we are going to check if the column scrapp data is in a date format 
# will also be necessary to group by  and use conditionals since we also be provinding data based by the months we have
# shows that we have an object so alright! that will work but we will make a little adjustment taking the days out, since are not valuable

# so i went and created new column with year and month only 

df['last_scraped_month'] = pd.to_datetime(df['last_scraped']).dt.strftime('%Y-%m')



```

```python
df[['price', 'minimum_nights', 'price_monthly','sq_meters','listing_url','price_sqm_amonth','last_scraped', 'last_scraped_month' ]]
```
