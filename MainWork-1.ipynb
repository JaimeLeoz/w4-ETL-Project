```python
#here im replacing the $ sign for nothin and also the "," in the thousands which was not letting me to change the column from a string to a float
# this way now i'm able do the next calculations without the error " cant multiply a string vs integer"
df['price'] = df['price'].str.replace('$','')
df['price'] = df['price'].str.replace(',','')
print(df.price)
```

```python
# confirming that has worked
print(df.dtypes)
```

```python
# calculations are fine and consistent

df[['price', 'minimum_nights', 'price_monthly','sq_meters','listing_url']]
```

```python
df.shape
```

```python
# we need to change the column sqm into a integer befor calculating anything
df["sq_meters"] = pd.to_numeric(df["sq_meters"], downcast="integer")
```

```python
# Here i do a loop to find all the elements of the rows, for pd dataframe. Im glad it worked.

tabla=driver.find_element(By.CLASS_NAME,'MuiTableBody-root')

filas=tabla.find_elements(By.CLASS_NAME,'MuiTableRow-root')


data=[]

for f in filas:
    
    elementos=f.find_elements(By.TAG_NAME,'td')
    
    tmp=[]
    
    for e in elementos:
        
        tmp.append(e.text)
        
    data.append(tmp)
    
data
```

```python
cabeceras =driver.find_element(By.CLASS_NAME,'MuiTableHead-root')

cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME,'th')]


cabeceras
```

```python
df_rent_municipio=pd.DataFrame(data, columns=cabeceras)

df_rent_municipio.head()
```
